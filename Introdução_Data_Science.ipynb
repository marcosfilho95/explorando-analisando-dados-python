```python
total_por_lingua = tmdb['original_language'].value_counts()
total_geral = total_por_lingua.sum()
total_de_ingles = total_por_lingua.loc['en']
total_do_rest_ = total_geral - total_de_ingles
print(total_geral, total_de_ingles, total_do_rest_)
```

```python
dados = {
  "lingua" : ["ingles", "outros"],
  "total" : [total_de_ingles, total_do_rest_]
}
dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(data = dados, x="lingua", y="total")
```

```python
dados.plot(kind="pie", labels=dados["lingua"], y="total")
```

```python
total_de_outros_filmes_por_lingua = tmdb.query("original_language != 'en'")["original_language"].value_counts()
total_de_outros_filmes_por_lingua.head()
```

# Comparar duas distribuições visualmente ou através de medidas

```python
filmes.head(2)
```

```python
notas_do_toy_story = notas.query("filmeId==1")["nota"]
notas_do_jumanji = notas.query("filmeId==2")["nota"]

media_toy_story = notas_do_toy_story.mean()
media_jumanji = notas_do_jumanji.mean()

print(media_toy_story, media_jumanji)

```

```python
mediana_toy_story = notas_do_toy_story.median()
mediana_jumanji = notas_do_jumanji.median()

print(mediana_toy_story, mediana_jumanji)
```

```python
import numpy as np

filme1 = [2.5]*10 + [3.5]*10
filme2 = [5]*10 + [1]*10

filme1_media = np.mean(filme1)
filme2_media = np.mean(filme2)

filme1_mediana = np.median(filme1)
filme2_mediana = np.median(filme2)

print("Filme 1:")
print(f"Media: {filme1_media}")
print(f"Mediana: {filme1_mediana}")

print("\nFilme 2:")
print(f"Media: {filme2_media}")
print(f"Mediana: {filme2_mediana}")
```

```python
plt.hist(filme1)
plt.hist(filme2)
plt.legend(["Filme 1", "Filme 2"])
```

```python
sns.boxplot((filme1, filme2))
```

```python
filme0 = [3.0]*20 # nao tem espalhamento
np.mean(filme0)
```
